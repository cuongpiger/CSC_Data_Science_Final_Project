```python
from modules.shopee_crawler import getAllRevirewsOfProduct, getAllRevirewsOfProduct2
import pandas as pd
```

```python
def readerProductsURLs(pFilePath):
    products_urls = None
    with open(pFilePath, "r") as reader:
        products_urls = [line.strip() for line in reader.readlines()]

    return products_urls
```

```python
def writeToCsv(pFilePath, pReviews):
    if pReviews == []: return 0
    reviews = []

    for row in pReviews:
        try:
            comment = row.iComment.strip()
            if comment == "" or comment == "****** Đánh giá đã bị ẩn do nội dung không phù hợp. ******": continue

            reviews.append((comment, row.iRating))
        except:
            continue

    df = pd.DataFrame(reviews, columns=['comment', 'rating'])
    df.to_csv(pFilePath, index = False, header=True)


    return df.shape[0]
```

```python
shopee_more_products_urls = readerProductsURLs("./data/shopee_products_urls/shopee_more_products_urls.txt")
```

```python
tmp = 0
no_reviews = 0

for i, url in enumerate(shopee_more_products_urls[tmp:]):
    reviews = getAllRevirewsOfProduct(url)
    file_path = "./data/shopee_reviews/more/product_{}.csv".format(tmp + i + 1)
    no_reviews += writeToCsv(file_path, reviews)

    print(no_reviews)
```
