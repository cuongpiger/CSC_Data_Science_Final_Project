```python
from modules.shopee_crawler import getAllRevirewsOfProduct
import pandas as pd
```

```python
def readerProductsURLs(pFilePath):
    products_urls = None
    with open("./data/shopee_products_urls/shopee_men_products_urls.txt", "r") as reader:
        products_urls = [line.strip() for line in reader.readlines()]

    return products_urls
```

```python
def writeToCsv(pFilePath, pReviews):
    if pReviews == []: return 0
    reviews = []

    for row in pReviews:
        try:
            comment = row.iComment.strip()
            if comment == "" or comment == "****** Đánh giá đã bị ẩn do nội dung không phù hợp. ******": continue

            reviews.append((comment, row.iRating))
        except:
            continue

    df = pd.DataFrame(reviews, columns=['comment', 'rating'])
    df.to_csv(pFilePath, index = False, header=True)


    return df.shape[0]
```

# 1. Xử lí cho các URL sản phẩm nam

```python
shopee_men_products_urls = readerProductsURLs("./data/shopee_products_urls/shopee_men_products_urls.txt")
```

```python
shopee_men_products_urls[:5]    
```

```python
no_reviews = 0

for i, url in enumerate(shopee_men_products_urls):
    reviews = getAllRevirewsOfProduct(url)
    file_path = "./data/shopee_reviews/men/product_{}.csv".format(i + 1)
    no_reviews += writeToCsv(file_path, reviews)

    print(no_reviews)
```
